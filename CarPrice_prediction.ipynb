# **Simple Linear Regression**

```python
X=df['enginesize'].values.reshape(-1,1)
X.shape
```

```python
y=df['price']
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.3,random_state=42)
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

### Check the how accurate data set

```python
lr.score(X_train, y_train)
```

```python
lr.score(X_test, y_test)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
#To draw best fit line on train data
plt.figure(figsize=(12,10));
sns.regplot(X_train,y_train);
plt.xlabel('Enginesize')
plt.ylabel("Price")
```

```python
lr.predict(np.array([100]).reshape(1,-1))
```

```python
y_train_pred= lr.predict(X_train)
y_test_pred=lr.predict(X_test)
```

### **Residual Analysis**

```python
# R2
from sklearn.metrics import r2_score
Train_Result=r2_score(y_train,y_train_pred)
Train_Result
```

```python
Test_Result=r2_score(y_test,y_test_pred)
Test_Result
```

# Data Standardization

```python
X=df.drop('price',axis=1)
y=df['price']
```

```python
from sklearn.preprocessing import StandardScaler
# define StandScaler scaler
scaler = StandardScaler()
# transform data
scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(scaled,y,test_size=0.2,random_state=42)
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(X_train, y_train)
Reg_predict=reg.predict(X_test)
```

```python
reg.score(X_train, y_train)
```

```python
reg.score(X_test, y_test)
```

```python
from sklearn import metrics
print("R score",metrics.r2_score(y_test,Reg_predict))
print("MSE = ", metrics.mean_squared_error(y_test,Reg_predict))
```

# Grid Search

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score

lm = LinearRegression()
scores = cross_val_score(lm, X_train, y_train, scoring='r2', cv=10)
print('scores',scores)
accuracy_regfold=np.mean(abs(scores))
print('\nCV mean',accuracy_regfold)
```

```python
from sklearn.feature_selection import RFE
# step-1: create a cross-validation scheme
folds = KFold(n_splits = 10, shuffle = True, random_state = 100)

# step-2: specify range of hyperparameters to tune
hyper_params = [{'n_features_to_select': list(range(1, 14))}]


# step-3: perform grid search
# 3.1 specify model
lm = LinearRegression()
lm.fit(X_train, y_train)
rfe = RFE(lm)      
```

```python
from sklearn.model_selection import GridSearchCV
model_cv = GridSearchCV(estimator = rfe, 
                        param_grid = hyper_params, 
                        scoring= 'r2', 
                        cv = folds, 
                        verbose = 1,
                        return_train_score=True)      

# fit the model
model_cv.fit(X_train, y_train) 
```

```python
model_cv.best_estimator_
```

```python
# final model
n_features_optimal = 13

lm = LinearRegression()
lm.fit(X_train, y_train)

rfe = RFE(lm, n_features_to_select=n_features_optimal)             
rfe = rfe.fit(X_train, y_train)

# predict prices of X_test
y_pred = lm.predict(X_test)
import sklearn.metrics
r2 = sklearn.metrics.r2_score(y_test, y_pred)
print(r2)
```
